**Task 6**

```python
A = np.random.randint(0, 100, size=150)
print("Random array of 150 integers:\n", A)

even_values = A[A % 2 == 0]
print("\nEven values from the array:\n", even_values)
num_even = len(even_values)
print("Number of even values:", num_even)

odd_values = A[A % 2 != 0]
print("\nOdd values from the array:\n", odd_values)
num_odd = len(odd_values)
print("Number of odd values:", num_odd)

def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

prime_values = A[np.vectorize(is_prime)(A)]

print("\nPrime values from the array:\n", prime_values)
print("Number of prime values:", len(prime_values))


def print_multiples_from_array(n, array):
  print(f"Multiples of {n} found in the array:")
  multiples_in_array = array[array % n == 0]
  if len(multiples_in_array) > 0:
    for multiple in multiples_in_array:
      print(multiple)
  else:
    print(f"No multiples of {n} found in the array.")
print_multiples_from_array(5, A)
```
