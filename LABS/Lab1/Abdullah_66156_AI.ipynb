```python
#Write a program to check if a number is even or odd

random_num = int(input("Enter a random number "))

if random_num % 2 == 0:
  print("The number is Even")
else :
  print("The number is Odd")
```

```python
#Find the largest of three numbers using conditions

num1 = int(input("Enter the first number "))
num2 = int(input("Enter the second number "))
num3 = int(input("Enter the third number "))

if num1 >= num2 and num1 >= num3 :
  largest = num1
elif num2 >= num1 and num2 >= num3 :
  largest = num2
else :
  largest = num3

print("The Largest number is " ,largest)

```

```python
#Print the first 10 terms of the Fibonacci sequence

terms = 10
a, b = 0, 1

print("first 10 terms of the Fibonacci sequence")

for x in range(terms):
  print(a)
  a, b = b, a + b
```

```python
#Write a program to calculate the factorial of a number using a loop

number = int(input("Enter the number "))
fact = 1

for i in range(1, number + 1):
  fact = fact * i

fact
```

```python
#Print table of the given number

number = int(input("Enter a number "))

for x in range(1, 11):
  print(number, "x", x, number * x)
```

```python
#task 5
#Count Digits in Number and their Sum

number1 =  input("Enter a number ")
count = 0
sumDigits = 0

for d in number1:
  count = count + 1
  sumDigits = sumDigits + int(d)

print("Number of digits in ", number1, "is", count)
print("Sum of digits is ", sumDigits)

```
