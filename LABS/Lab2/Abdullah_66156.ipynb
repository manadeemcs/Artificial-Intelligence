**Task 1**
**Appending the List**

```python
vegetables = ["potato", "onion", "cucumber"]
vegetables.append("lady finger")
print(vegetables)
```

**Task 2**
**Clearing the List**

```python
car_brand = ["Mercedes", "BMW", "Porsche"]
car_brand.clear()
print(car_brand)
```

**Task 3**
**Copying the List**

```python
car_brand = ["Mercedes", "BMW", "Audi"]
print(car_brand)
car_brand2 = car_brand.copy()
print(car_brand2)
car_brand2.append("porsche")
print(car_brand2)
```

**Task 4**
**Counting the List**

```python
jets = ["F-16", "F-35 raptor", "B2 Bomber", "JF-17 Thunder", "F-16", "F-35 raptor", "F-16"]
print(jets)
print(jets.count("F-16"))
print(jets.count("F-35 raptor"))
```

**Task 5**
**Extending the List**

```python
#extend 5
vegetables = ["potato", "onion", "cucumber"]
fruits = ["coconut", "pine apple", "kiwi"]
print(vegetables)
print(fruits)
fruits.extend(vegetables)
print(fruits)
```

**Task 6**
**Indexing the List**

```python
car_brand = ["Mercedes", "BMW", "Audi", "Porsche", "Honda", "Mazda", "Nissan"]
print(car_brand)
print(car_brand.index("BMW"))
print(car_brand.index("Porsche"))
# negative indexing
print(car_brand[1:4])
print(car_brand[-5:-2])

```

**Task 7**
**Inserting the List**

```python
car_brand = ["Mercedes", "BMW", "Audi", "Porsche", "Honda", "Mazda", "Nissan"]
print(car_brand)
car_brand.insert(5,"Skoda")
print(car_brand)

```

**Task 8**
**Poping the List**

```python
car_brand = ["Mercedes", "BMW", "Audi", "Porsche", "Honda", "Mazda", "Nissan"]
print(car_brand)
car_brand.pop(3)
print(car_brand)
car_brand.pop(3)
print(car_brand)
```

**Task 9**
**Removing the List**

```python
car_brand = ["Mercedes", "BMW", "Audi", "Porsche", "Honda", "Mazda", "Nissan"]
print(car_brand)
car_brand.remove("Audi")
print(car_brand)
car_brand.remove("Porsche")
print(car_brand)
```

**Task 10**
**Reversing the List**

```python
car_brand = ["Mercedes", "BMW", "Audi", "Porsche", "Honda", "Mazda", "Nissan"]
print(car_brand)
car_brand.reverse()
print(car_brand)
```

**Task 11**
**Sorting the List**

```python
car_brand = ["Mercedes", "BMW", "Audi", "Porsche", "Honda", "Mazda", "Nissan"]
print(car_brand)
car_brand.sort()
print(car_brand)

numaric_list = [2, 434, 1, 3, 5, 0]
print("Numeric list sorting in ascending order")
print(numaric_list)
numaric_list.sort()
print(numaric_list)

print("Numeric list sorting in descending order")
numaric_list.sort(reverse=True)
print(numaric_list)
```

BONUS


```python
fruit_list = ['apple', 'banana', 'apple', 'orange', 'banana', 'apple']

counts = {}

for item in fruit_list:
    if item in counts:
        counts[item] += 1
    else:
        counts[item] = 1

print(counts)
```
